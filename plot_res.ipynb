```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

colors = sns.color_palette("tab10")
```

# Hydrogen

```python
H2_50 = np.array([10134476434058, 8715727146582, 302163614894, 267402005934, 254061616754, 558354851496])
H2_100 = np.array([23756679740822, 17169260922146, 450117916330, 450561598598, 514393952386, 1158344702628])
H2_200 = np.array([51316236673082, 36990622498072, 1243605496466, 1114613230262, 1058896775912, 274222809830])

exact_H2 =  2 * 11994**3

factor = np.array([0.7, 1, 1.5, 2, 2.5, 3])

N = np.array([50, 100, 200])
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4), sharey=True)


x_ticks0 = factor * N[0]
x_ticks_labels = [f"{x:.1f}" for x in factor]
axes[0].plot(x_ticks0, H2_50 / 1e12, marker='o', label='N=50', color=colors[0])
axes[0].plot(x_ticks0, [exact_H2 / 1e12]*len(x_ticks0), linestyle='--', label='Exact', color='black')
axes[0].set_yscale('log')
axes[0].set_xlabel('Naux / N')
axes[0].set_title('N=50')
axes[0].set_xticks(x_ticks0, labels=x_ticks_labels)

x_ticks1 = factor * N[1]
axes[1].plot(x_ticks1, H2_100 / 1e12, marker='^', label='N=100', color=colors[1])
axes[1].plot(x_ticks1, [exact_H2 / 1e12]*len(x_ticks1), linestyle='--', label='Exact', color='black')
axes[1].set_yscale('log')
axes[1].set_xlabel('Naux / N')
axes[1].set_title('N=100')
axes[1].set_xticks(x_ticks1, labels=x_ticks_labels)

x_ticks2 = factor * N[2]
axes[2].plot(x_ticks2, H2_200 / 1e12, marker='x', label='N=200', color=colors[2])
axes[2].plot(x_ticks2, [exact_H2 / 1e12]*len(x_ticks2), linestyle='--', label='Exact', color='black')
axes[2].set_yscale('log')
axes[2].set_xlabel('Naux / N')
axes[2].set_title('N=200')
axes[2].set_xticks(x_ticks2, labels=x_ticks_labels)

axes[0].set_ylabel(r'Flops ($\times 1e12$)')
plt.tight_layout()
```

# Formaldehyde

```python
factor = np.array([2, 2.5, 3])

formaldehyde_50 = np.array([16242097824685, 13368249830289, 18965919642037])
formaldehyde_100 = np.array([22348217655421, 21934551468586, 20214942741175])
formaldehyde_200 = np.array([33487479213871, 37366877152400, 33201057403130])

exact_formaldehyde =  20 * 27643**3
```
